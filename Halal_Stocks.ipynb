# ১. লিকুইড অ্যাসেটের অনুপাত (%)


```python
def liquid_asset_ratio_with_rating(liquid_assets, total_assets):
    ratio = (liquid_assets / total_assets) * 100
    if ratio >= 50:
        rating = "⭐⭐⭐⭐⭐ - Excellent liquidity!"
    elif ratio >= 40:
        rating = "⭐⭐⭐⭐ - Good liquidity."
    elif ratio >= 30:
        rating = "⭐⭐⭐ - Moderate liquidity."
    elif ratio >= 20:
        rating = "⭐⭐ - Low liquidity."
    else:
        rating = "⭐ - Poor liquidity."
    return ratio, rating

# উদাহরণ
liquid_ratio, liquid_rating = liquid_asset_ratio_with_rating(liquid_assets=100000, total_assets=200000)
print(f"লিকুইড অ্যাসেট অনুপাত: {liquid_ratio:.2f}% | রেটিং: {liquid_rating}")
```

# ২. ঋণের অনুপাত (Debt Ratio)

```python
def debt_ratio_with_rating(total_debt, total_assets):
    ratio = (total_debt / total_assets) * 100
    if ratio <= 20:
        rating = "⭐⭐⭐⭐⭐ - Excellent debt management!"
    elif ratio <= 30:
        rating = "⭐⭐⭐⭐ - Good debt control."
    elif ratio <= 40:
        rating = "⭐⭐⭐ - Acceptable debt level."
    elif ratio <= 50:
        rating = "⭐⭐ - High debt level."
    else:
        rating = "⭐ - Poor debt management."
    return ratio, rating

# উদাহরণ
debt_ratio_result, debt_rating = debt_ratio_with_rating(total_debt=60000, total_assets=200000)
print(f"ঋণের অনুপাত: {debt_ratio_result:.2f}% | রেটিং: {debt_rating}")
```

# ৩. হারাম আয়ের অনুপাত (Non-Compliant Income Ratio)

```python
def non_compliant_income_ratio_with_rating(non_compliant_income, total_income):
    ratio = (non_compliant_income / total_income) * 100
    if ratio <= 5:
        rating = "⭐⭐⭐⭐⭐ - Excellent compliance with Sharia!"
    elif ratio <= 10:
        rating = "⭐⭐⭐⭐ - Good compliance."
    elif ratio <= 15:
        rating = "⭐⭐⭐ - Moderate compliance."
    elif ratio <= 20:
        rating = "⭐⭐ - Below average compliance."
    else:
        rating = "⭐ - Poor compliance."
    return ratio, rating

# উদাহরণ
non_compliant_ratio, non_compliant_rating = non_compliant_income_ratio_with_rating(non_compliant_income=3000, total_income=100000)
print(f"হারাম আয়ের অনুপাত: {non_compliant_ratio:.2f}% | রেটিং: {non_compliant_rating}")
```

# ৪. ইন্টারেস্ট-বেয়ারিং ঋণের অনুপাত (%)

# ৫. সম্পদের আয় থেকে আয়ের অনুপাত (Income from Assets Ratio)

```python
def income_from_assets_ratio_with_rating(income_from_assets, total_assets):
    ratio = (income_from_assets / total_assets) * 100
    if ratio >= 10:
        rating = "⭐⭐⭐⭐⭐ - Excellent return on assets!"
    elif ratio >= 7:
        rating = "⭐⭐⭐⭐ - Good return on assets."
    elif ratio >= 5:
        rating = "⭐⭐⭐ - Moderate return on assets."
    elif ratio >= 3:
        rating = "⭐⭐ - Below average return."
    else:
        rating = "⭐ - Poor return."
    return ratio, rating

# উদাহরণ
income_ratio, income_rating = income_from_assets_ratio_with_rating(income_from_assets=15000, total_assets=200000)
print(f"সম্পদের আয় থেকে আয়ের অনুপাত: {income_ratio:.2f}% | রেটিং: {income_rating}")
```

# ৬. মুনাফা (Profit Margin)

```python
def profit_margin_with_rating(profit, revenue):
    margin = (profit / revenue) * 100
    if margin >= 20:
        rating = "⭐⭐⭐⭐⭐ - Excellent profitability!"
    elif margin >= 15:
        rating = "⭐⭐⭐⭐ - Good profitability."
    elif margin >= 10:
        rating = "⭐⭐⭐ - Moderate profitability."
    elif margin >= 5:
        rating = "⭐⭐ - Below average profitability."
    else:
        rating = "⭐ - Poor profitability."
    return margin, rating

# উদাহরণ
profit_margin_result, profit_margin_rating = profit_margin_with_rating(profit=25000, revenue=100000)
print(f"মুনাফা মার্জিন: {profit_margin_result:.2f}% | রেটিং: {profit_margin_rating}")
```

# ৭. পরিশোধযোগ্যতা অনুপাত (Liquidity Ratio)

```python
def liquidity_ratio_with_rating(current_assets, current_liabilities):
    ratio = current_assets / current_liabilities
    if ratio >= 3:
        rating = "⭐⭐⭐⭐⭐ - Excellent liquidity."
    elif ratio >= 2:
        rating = "⭐⭐⭐⭐ - Good liquidity."
    elif ratio >= 1:
        rating = "⭐⭐⭐ - Moderate liquidity."
    elif ratio >= 0.5:
        rating = "⭐⭐ - Low liquidity."
    else:
        rating = "⭐ - Poor liquidity."
    return ratio, rating

# উদাহরণ
liquidity_ratio_result, liquidity_ratio_rating = liquidity_ratio_with_rating(current_assets=150000, current_liabilities=50000)
print(f"পরিশোধযোগ্যতা অনুপাত: {liquidity_ratio_result:.2f} | রেটিং: {liquidity_ratio_rating}")
```

# Liquidity
If a company's liquidity ratio is very high (e.g., 3 or more), it can be considered compliant with Sharia principles. However, excessive liquidity might sometimes indicate lower economic efficiency.

# Income from Assets Ratio
If a company earns sufficient income from its assets (e.g., more than 10%-15%), it can be deemed halal, as excessive debt or interest-based income goes against Sharia principles.

# Profit Margin
If the profit margin exceeds 10%-15%, it is considered good under Sharia, as it reflects proper management of the company's operations and income. Excessive profit margins or corporate greed can sometimes be considered haram.

# Islamic Business Model
The type of business the company engages in is crucial. If the company is involved in interest-based activities, alcohol, or gambling-related businesses, it is considered haram, even if other financial criteria are favorable.

# Debt-Related Issues
If a company has significant debt (e.g., more than 30%-50%), it goes against Sharia, as Islam prohibits interest (Riba) on loans. A lower level of debt makes a company more likely to be halal.

# Liquid Assets Balance
If a company maintains a proper balance of liquid assets (e.g., cash, stocks), it is considered favorable. Too little or too much liquid assets balance is not compliant with Sharia.

# Sharia-Compliant Investments
The nature of a company’s investments must be verified. If the company invests in interest-based, volatile, or impermissible businesses or sectors, it may be deemed haram. In this case, reviewing the company’s investment profile is essential.

# ১. লিকুইডিটি (Liquidity)
- যদি কোম্পানির লিকুইডিটি অনুপাত (Liquidity Ratio) খুব বেশি থাকে (যেমন ৩ বা তার বেশি), তাহলে এটি শারিয়া অনুসারে সঠিক বলে ধরা যেতে পারে। তবে, খুব বেশি লিকুইডিটি থাকলে, এটিও কখনও কখনও অর্থনৈতিক কার্যকারিতা কম থাকতে পারে।
# ২. আয় থেকে আয়ের অনুপাত (Income from Assets Ratio)
- যদি কোম্পানি তার সম্পদ থেকে যথেষ্ট আয় অর্জন করে (যেমন ১০%-১৫% এর বেশি), তবে এটি হালাল বলে বিবেচিত হতে পারে, কারণ অতিরিক্ত ঋণগ্রস্ততা বা সুদভিত্তিক আয় শরিয়ার বিরুদ্ধে যায়।
# ৩. মুনাফা মার্জিন (Profit Margin)
- মুনাফা মার্জিন যদি ১০%-১৫% এর বেশি থাকে, তাহলে এটি শরিয়া অনুযায়ী ভাল হবে, কারণ এটি কোম্পানির কর্মক্ষমতা ও তার আয় সঠিকভাবে পরিচালিত হচ্ছে তা প্রকাশ করে। অতিরিক্ত মুনাফা বা অতিরিক্ত কর্পোরেট মুনাফা কখনও কখনও হারাম হতে পারে।
# ৪. শুধুমাত্র ইসলামিক ব্যবসা (Islamic Business Model)
- কোম্পানির ব্যবসার ধরন গুরুত্বপূর্ণ। যদি কোম্পানি সুদভিত্তিক কার্যক্রম, মদ, বা জুয়া সম্পর্কিত ব্যবসা করে, তবে তা হারাম হতে পারে, তা সত্ত্বেও যদি তার অন্যান্য আর্থিক মানদণ্ড ভালো হয়।
# ৫. ঋণ-সম্পর্কিত বিষয় (Debt-related issues)
- যদি কোম্পানির বেশি ঋণ থাকে (যেমন কোম্পানি ৩০%-৫০% এর বেশি ঋণ নিয়ে চলতে থাকে), তবে এটি শরিয়া বিরুদ্ধে যাবে, কারণ ইসলামে সুদ (Riba) গ্রহণ বা প্রদান হারাম। ঋণ-সম্পর্কিত বিষয়গুলো কম হলে এটি হালাল বলে ধরা যেতে পারে।
# ৬. লিকুইড এসেটের ভারসাম্য (Liquid Assets Balance)
- যদি কোম্পানির লিকুইড এসেট (যেমন নগদ, স্টক) তুলনামূলকভাবে সঠিক ভারসাম্য বজায় রাখে, তাহলে এটি ভালো। খুব কম বা অত্যধিক লিকুইড এসেটের ভারসাম্য শারিয়া অনুকূল নয়।
# ৭. ইসলামিক বিনিয়োগ (Sharia-Compliant Investments)
- কোম্পানির বিনিয়োগের ধরন নিশ্চিত করতে হবে। যদি তারা সুদভিত্তিক, অস্থির বা অশুদ্ধ ব্যবসা বা সেক্টরে বিনিয়োগ করে, তাহলে সেটা হারাম হতে পারে। এক্ষেত্রে, কোম্পানির বিনিয়োগ প্রোফাইল পরীক্ষা করতে হবে।
